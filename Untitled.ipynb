```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/covid_19_data.csv")
```

```python
df = df.replace(np.nan, "dhoka", regex=True)
df
```

```python
cases_by_location = df.groupby(['Province/State', 'Country/Region']).tail(1)
cases_by_location.head(100)
```

```python
data = {}
for location in cases_by_location.iterrows():
    print(location[1]["Deaths"])
#     print(name)
    string = location[1]["Province/State"].split(",")[0] if location[1]["Province/State"] != "dhoka" else location[1]["Country/Region"]    
    data[string] = {}
    data[string]["Confirmed"] = location[1]["Confirmed"]
    data[string]["Recovered"] = location[1]["Recovered"]
    data[string]["Deaths"] = location[1]["Deaths"]
    
# print(data)
```

```python
data
```

```python
import requests 
import urllib
```

```python
for key, value in data.items():
    print(key)
    URL = "http://api.geonames.org/searchJSON?q={}&maxRows=1&username=sanchittanwar7".format(urllib.quote(key))
#     print(URL)
    r = requests.get(url = URL) 

    # extracting data in json format 
    output = r.json()
    if len(output['geonames']) >= 1:
        value["lat"] = output['geonames'][0]['lat'].encode("utf-8")
        value["lng"] = output['geonames'][0]['lng'].encode("utf-8")

print(data)
```
